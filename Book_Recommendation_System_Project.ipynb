```python
# prompt: generate pivot table

pivot_table1=df.pivot_table(values='Approximate sales in millions', index='Genre', columns='Book',fill_value=0)
pivot_table1
```

```python
X = pivot_table1.T
```

```python
X
```

```python
import sklearn
from sklearn.decomposition import TruncatedSVD
```

```python
SVD = TruncatedSVD(n_components=10) # This line initializes an SVD object with n_components set to 10. n_components specifies the number of components to retain after the decomposition. In this case, it will retain the 10 most significant components.
decomposed_matrix = SVD.fit_transform(X)
decomposed_matrix.shape
```

```python
correlation_matrix = np.corrcoef(decomposed_matrix)
correlation_matrix.shape
```

```python
X.index[55]
```

```python
i = "She: A History of Adventure"

book_name = list(X.index)
book_name = book_name.index(i)
book_name
```

```python
correlation_book_name = correlation_matrix[book_name]
correlation_book_name.shape
```

```python
Recommend = list(X.index[correlation_book_name > 0.90])

# Removes the item already bought by the customer
Recommend.remove(i)

Recommend[0:9]
```

```python
# prompt: apply the cosine function as above to our project too

# Create a user-item matrix using pivot table
user_item_matrix = df.pivot_table(index='Book', columns='Genre', values='Approximate sales in millions', fill_value=0)

# Calculate the cosine similarity between users
cosine_sim = sklearn.metrics.pairwise.cosine_similarity(user_item_matrix)

# Get the index of the item the user is interested in
item_index = user_item_matrix.index.get_loc("Harry Potter and the Philosopher's Stone")

# Get the similarity scores for the item
similarity_scores = cosine_sim[item_index]

# Get the top 10 most similar items
similar_items = user_item_matrix.index[similarity_scores.argsort()[-10:]]

# Print the similar items
print("Similar Items:")
print(similar_items)
```
